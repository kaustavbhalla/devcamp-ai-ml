### Step 2: Monthly Analysis
Monthly analysis of various columns across the data

```python
df['date'] = pd.to_datetime(df['date'])
df['date'].dtype
```

<-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------->

#### January

```python
janDf = df[df['date'].dt.month == 1]
```

```python
janDf
```

Workout Type

```python
print(janDf['workout_type'].value_counts())
```

Water Intake(In Litres)

```python
print(f"{janDf['water_intake_liters'].sum():.2f}")
```

Distance Traversed(In Kilometres)

```python
print(janDf['distance_km'].sum())
```

Calories Burned

```python
print(janDf['calories_burned'].sum())
```

Finding Correlations

```python
janDf.select_dtypes(include=['number']).corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(janDf.select_dtypes(include=['number']).corr(), annot=True)
```
